```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.float_format', '{:.2f}'.format)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
table = pd.read_excel('lr1.xlsx')
```

```python
# Исходные данные 12-го варианта
table12 = pd.concat([
    table[7:11],
    table[12:25],
    table[32:37],
    table[38:46],
    table[72:78],
    table[79:],
])
table12.head()
```

```python
# Отделение функции от параметров
xs = table12.iloc[:,2:]
y = table12.iloc[:,1]
xs.plot()
```

```python
# Нормализация данных
normal = (xs - xs.mean())/xs.std()
normal.plot()
```

```python
normal.head()
```

```python
# Вычислим псевдообратную матрицу
normal.insert(0, 'x0', 1.0)
theta = np.linalg.inv(normal.T.dot(normal)).dot(normal.T).dot(y)

# Вывод весов регрессии
print(pd.DataFrame(theta, columns=['theta']))
```

```python
# Корреляционная матрица
corr_matrix = pd.DataFrame(np.corrcoef(normal.drop(columns = ['x0']).T))
corr_matrix.style.background_gradient(cmap='Greens')
```

```python
# Число обусловленности < 10^1. Матрица хорошо обусловалена. Частичная Мультиколлинеарность достаточно высока.
np.linalg.cond(normal.drop(columns = ['x0']))
```

```python
# Рассчет ошибки линейной модели
_y = np.matmul(normal, theta)
E = y - _y
MSE_A = 1/len(E)*sum(E**2)
R2_A = r2_score(y, _y)
print(MSE_A)
```

```python
from scipy.stats import norm
default = np.arange(-5, 5, 0.001)
E.plot.kde().plot(default, norm.pdf(default, 0, 1), color='red')
```

```python
# Определение весов линейной регрессии
# Начальные веса
w = np.array( [1]*10 )

# Скорость обучения
lr = 0.01

# Количество итераций
n_iter = 10000

# MSE и MSE'
cost = lambda y, _y: (1/len(y)) * np.sum((_y - y)**2)
cost_derivative = lambda x, y, _y: (2/len(y)) * np.matmul(-x.T, (y - _y))

# Градиентный спуск
errors = np.empty(0)
for i in range(n_iter):
    _y = np.matmul(normal, w)
    errors = np.append(errors, cost(y, _y))
    w -= lr * cost_derivative(normal, y, _y)

MSE_GM = errors[-1]
R2_GM = r2_score(y, _y)

# Вывод на экран графика изменения ошибки
pd.DataFrame(errors, columns=['errors']).plot(logy=True, logx=True)
```

```python
# Сравнение результатов Аналитического и градиентного методов.
print('Результат Аналитического метода:' + str(MSE_A).rjust(20))
print('Результат Градиентного метода:' + str(MSE_GM).rjust(22))
```

```python
# Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(normal, y, test_size=0.2, random_state=42)
# Создание и обучение модели линейной регрессии
model = LinearRegression()
model.fit(X_train, y_train)

# Прогнозирование на тестовой выборке
y_pred = model.predict(X_test)

MSE_S = cost(y, _y)
R2_S = r2_score(y, _y)
print("Результат sklearn:", MSE_S, R2_S)
```
